#Swapping Numbers Without third variable


```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Before swapping: a =", a, ", b =", b)

a = a + b  # a becomes sum of a and b
b = a - b  # b becomes original a (a+b - b)
a = a - b  # a becomes original b (a+b - original a)

print("After swapping: a =", a, ", b =", b)
```

#Create a student Report Program that take student details using input(),Store marks in variables ,Calculate total and percentage, use comments , use proper indentation


```python
name = input("Student Name:")
s_class = input("Class:")
enrollment_no = int(input("Enrollment Number:"))

# Taking marks input and storing them in temporary variables
dsa_marks = float(input("DSA Marks:"))
ids_marks = float(input("IDS Marks:"))
m_marks = float(input("M Marks:"))
p_marks = float(input("P Marks:"))
evs_marks = float(input("EVS Marks:"))

# Calculate Total and Percentage using the temporary variables
total_marks = dsa_marks + ids_marks + m_marks + p_marks + evs_marks
percentage = (total_marks / 500) * 100 # Assuming each subject is out of 100, so 5 subjects total 500

Student_report = {
    'Name': name,
    'S_Class': s_class,
    'Enrollment_no': enrollment_no,
    'DSA': dsa_marks,
    'IDS': ids_marks,
    'M': m_marks,
    'P': p_marks,
    'EVS': evs_marks,
    'Total': total_marks,
    'Percentage': percentage
}
print(Student_report)
```
