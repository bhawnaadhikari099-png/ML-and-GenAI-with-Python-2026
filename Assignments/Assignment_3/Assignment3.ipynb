#Calculate sum of first N natural numbers

```python
n = int(input("Enter number:"))

def Sum(n):
  total = 0
  for i in range(1,n+1):
       total+=i
  return total

result = Sum(n)
print("sum of the first n natural numbers :",result)
```

#Reverse a Number

```python
n = int(input("Enter Number:"))
def reverse(n):
  reversed_num = 0
  temp = n
  while temp > 0:
    digit = temp % 10
    reversed_num = reversed_num * 10 + digit
    temp = temp // 10
  return reversed_num

result = reverse(n)
print("Reversed number:", result)
```

#Count digits in a number

```python
n = int(input("Enter number:"))
def Count(n):
    count = 0
    temp = n
    if temp == 0: # Handle case for n=0
        return 1
    while temp > 0:
        temp = temp // 10
        count += 1
    return count

Total = Count(n)
print("Total digits:",Total)
```

#Palindrome number

```python
n = int(input("Enter number:"))

def Palindrome(num_to_check):
  # Nested helper function to reverse a number
  def reverse_helper(val):
    reversed_num = 0
    temp = val
    while temp > 0:
      digit = temp % 10
      reversed_num = reversed_num * 10 + digit
      temp = temp // 10
    return reversed_num
  reversed_val = reverse_helper(num_to_check)

  # Compare the original number with its reversed version
  if num_to_check == reversed_val:
    return True
  else:
    return False

result = Palindrome(n)
print(result)
```

#Fibonacci Series

```python
def fibonacci_iterative(n):
    series = []
    a, b = 0, 1
    for _ in range(n):
        series.append(a)
        a, b = b, a + b  # Simultaneous assignment updates both variables at once
    return series

print(fibonacci_iterative(8))
# Output: [0, 1, 1, 2, 3, 5, 8, 13]
```

#CALCULATOR

```python
def calculator():
    print("Select operation:")
    print("1.addition")
    print("2.substraction")
    print("3.multiplication")
    print("4.division")
    choice = int(input("Enter choice(1/2/3/4):"))
    num1 = float(input("Enter first number:"))
    num2 = float(input("Enter second number:"))

    if choice == 1:
        return num1 + num2
    elif choice == 2:
        return num1 - num2
    elif choice == 3:
        return num1 * num2
    elif choice == 4:
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
            return None
        return num1 / num2
    else:
        print("Invalid input. Please choose a number between 1 and 4.")
        return None

# Call the function and print the result.
# These lines should be at the top-level (no indentation).
result = calculator()
if result is not None:
    print("Result:", result)
```

#Handle division by 0 using exception handling


```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))

def divide(a,b):
  try:
    result = a/b
  except ZeroDivisionError:
     print("Error: Division by zero is not allowed.")
     return None
  else:
    return result

result = divide(num1,num2)
if result is not None:
  print("Result:",result)
```
