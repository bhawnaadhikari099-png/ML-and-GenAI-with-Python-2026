#Factorial of number


```python
n = int(input("Enter number:"))

fact =1
if(n<0):
  print("Negative number donot have factorial")
elif(n==0):
  print("Factorial of 0 : 1")
else:
  for i in range(n):
    fact = fact*(i+1)
  print("Factorial of",n,":",fact)




```

#Fibonacci series

```python
n = int(input("Enter number:"))

a, b = 0, 1

if n <= 0:
    print("Please enter a positive integer")
elif n == 1:
    print("Fibonacci sequence up to", n, ":")
    print(a)
else:
    print("Fibonacci sequence up to", n, ":")
    for _ in range(n):
        print(a)
        nth = a + b
        a = b
        b = nth
```

#Largest among 3 numbers

```python
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
c=int(input("Enter third number:"))

if(a>b and a>c):
  print(a,"is largest")
elif(b>a and b>c):
  print(b,"is largest")
else:
  print(c,"is largest")

```
